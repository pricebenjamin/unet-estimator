```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
images = np.load('eval_images.npy')
masks = np.load('eval_pred_masks.npy')

n = len(images)
print('Number of images: %s' % n)

fig, axes = plt.subplots(3, 2, figsize=(15, 20))

idx = np.random.choice(n, size=3, replace=False)

for i, (ax1, ax2) in enumerate(axes):
    ax1.imshow(images[idx[i]].astype(np.uint8))
    ax2.imshow(masks[idx[i]])
```
